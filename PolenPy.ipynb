```python
import requests #Sluzi sa slanje requestova ka API-ju
import os.path as io #Biblioteka za rad sa fajlovima
from datetime import date, timedelta #timedelta sluzi za oduzimanje/dodavanje datuma
import time
from colorama import Fore, Style #Ove biblioteke smo koristili da menjamo boju teksta za unos

print("--=[ PolenPy by Luka Kovacevic ]=--")
time.sleep(1)
print("--=[ Skripta pokazuje koncentraciju polena u odabranom gradu u poslednjih X dana. ]=--")
time.sleep(3)

print(Fore.GREEN)
days_num = input("Unesi broj dana: ") #Trazimo od korisnika da unese broj dana
print(Style.RESET_ALL)


today = date.today() #Uzimamo danasnji datum
#today = today.strftime("%d-%m-%Y") #Foramtiramo datum
starting_date = today - timedelta(3 + int(days_num)) #Oduzimamo 3 dana zato sto sam api kasni 3 dana i jos koliko dana je korisnik rekao da zeli da vidi unazad



if not io.isfile("locations.txt"): #Proveravamo da li fajl sa imonom "locations.txt" vec postoji
    file = open("locations.txt", "w", encoding="utf-8") #Kreiramo fajl pod imenom "location.txt", kazemo da radimo samo upisivanje "w", i da radi encoding = "utf-8"
    
    location_list = requests.get("http://polen.sepa.gov.rs/api/opendata/locations/") #Uzimamo podatke sa API-a sa zadate putanje
    location_list = location_list.json() #Podatke prebacujemo u JSON format

    location_list = sorted(location_list, key = lambda k: k['id']) #Sortiramo podatke po atributu id

    for item in location_list: #Prolazimo kroz location_list i upisujemo jedan po jedan red u novokreirani fajl
        file.write(f"{item['id']}: {item['name']} - {item['description']}\n")

    file.close()

locations = open("locations.txt", "r", encoding="utf-8") #Citamo fajl (Cita liniju po liniju)
# for line in locations:
#     print(line, end='') #end='' uklanja prazan prostor izmedju redova (Uklanja visak spaceova)

locations.close()

location = input("Unesi mesto (broj): ")

pollen = requests.get(f"http://polen.sepa.gov.rs/api/opendata/pollens/?location={location}&date_after={starting_date}")
pollen = pollen.json()

#print(pollen['results'])

pollen = pollen['results']

print("=========================================\n")

for item in pollen:
    print(f"Za dan: {Fore.BLUE}{item['date']}{Style.RESET_ALL}")
    for inner_item in item['concentrations']:
        concentrations = requests.get(f"http://polen.sepa.gov.rs/api/opendata/concentrations/{inner_item}/")
        concentrations = concentrations.json()
        
        allergen = requests.get(f"http://polen.sepa.gov.rs/api/opendata/allergens/{concentrations['allergen']}/")
        allergen = allergen.json()

        current_value = concentrations['value']

        if current_value > allergen['margine_top']:
            print(f"Koncentracija polena {allergen['localized_name']} ({allergen['name']}) je visoka!!!")
        elif allergen['margine_top'] >= current_value >= allergen['margine_bottom']:
            print(f"Koncentracija polena {allergen['localized_name']} ({allergen['name']}) je srednja!")
        else:
            print(f"Koncentracija polena {allergen['localized_name']} ({allergen['name']}) je niska.")
        
    print("=========================================\n")












```
